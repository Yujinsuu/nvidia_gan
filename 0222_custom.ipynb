```python
# 파일을 접근 -> 메모리에 파일을 올려야 -> 모든 파일의 경로가 필요
from glob import glob
mother_path = '/content/train'

glob(mother_path+'/*/*')
```

```python
import os
import numpy as np
from PIL import Image
import tensorflow as tf
from tensorflow.keras import layers, models

folders = ['Romance', 'Fantasy']

train = []
label = []
for idx , folder_name in enumerate(folders):
    folder = os.path.join(mother_path, folder_name)
    all_files_path = glob(folder+'/*')

    num = 0
    for file_path in all_files_path:
      num+=1

    for k,file_path in enumerate(all_files_path):
        # 파일 불러오기
        img = Image.open(file_path)
        # 이미지 파일 전처리
        img = img.resize((128,128))
        img = img.convert('RGB')
        np_img = np.array(img)
        train.append(np_img)
        label.append(idx)
        if k % 20 == 0:
            print(f'{k} / {len(all_files_path)}')
np_train = np.array(train)
np_label = np.array(label)
```

```python
print(np.bincount(np_label))
print(np_train.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(np_train, np_label, test_size=0.2)
```

```python
x_train = x_train/255.
x_test = x_test/255.
```

```python
input_layer = layers.Input(shape=(128,128,3), name='start layer')
start_layer = input_layer
for idx in range(2):
  tower_1 = layers.Conv2D(64 , kernel_size=(1,1) ,
                        padding='same', strides=(1,1),
                        activation='relu', name='tower'+str(idx+1)+'_1')(input_layer)
  tower_2 = layers.Conv2D(64 , kernel_size=(1,1) ,
                        padding='same', strides=(1,1),
                        activation='relu', name='tower'+str(idx+1)+'_2_1')(input_layer)
  tower_2 = layers.Conv2D(64 , kernel_size=(3,3) ,
                        padding='same', strides=(1,1),
                        activation='relu', name='tower'+str(idx+1)+'_2_2')(tower_2)
  tower_3 = layers.Conv2D(64 , kernel_size=(1,1) ,
                        padding='same', strides=(1,1),
                        activation='relu', name='tower'+str(idx+1)+'_3_1')(input_layer)
  tower_3 = layers.Conv2D(64 , kernel_size=(5,5) ,
                        padding='same', strides=(1,1),
                        activation='relu', name='tower'+str(idx+1)+'_3_2')(tower_3)
  tower_4 = layers.MaxPool2D(pool_size=(3,3) , strides=(1,1) ,
                           padding='same', name='tower'+str(idx+1)+'_4_1')(input_layer)
  tower_4 = layers.Conv2D(64 , kernel_size=(1,1) ,
                        padding='same', strides=(1,1),
                        activation='relu', name='tower'+str(idx+1)+'_4_2')(tower_4)
  concat_layer = layers.concatenate([tower_1,tower_2,tower_3,tower_4],axis=3,name='concat'+str(idx+1))
  input_layer = layers.MaxPool2D(pool_size=(3,3), strides=(2,2), padding='valid',
                                   name='concat'+str(idx+1)+'_maxpool')(concat_layer)

last_avg_pool = layers.AveragePooling2D(pool_size=(31,31) , strides=(1,1), padding='valid')(input_layer)
flat_layer = layers.Flatten()(last_avg_pool)
output_layer = layers.Dense(1 ,activation='sigmoid')(flat_layer)
model = models.Model(start_layer,output_layer)
model.summary()
```

```python
!mkdir tensorboard_log
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',metrics=['accuracy'])
tb = tf.keras.callbacks.TensorBoard(log_dir='/content/tensorboard_log',
                                    histogram_freq=1)
model.fit(x_train,y_train,
          validation_data=(x_test,y_test),
          epochs=10,callbacks=[tb])
```
